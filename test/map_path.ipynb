```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def read_file(path):
    with open(path, "r") as file:
        data = [line.split(";")[1:] for line in file.readlines()]
        data = [(float(line[0]), float(line[1])) for line in data]

    return data

points = read_file("data/meta1.txt")
```

```python
def calculate_dif(distance, angle):
    angle = np.deg2rad(angle)
    x =  distance * np.cos(angle)
    y =  distance * np.sin(angle)
    return x, y

path = []
x, y = 0,0
last_distance = 0
for (distance, angle) in points:
    x_dif, y_dif = calculate_dif(last_distance - distance, angle)
    last_distance = distance
    x += x_dif
    y += y_dif
    path.append((x, y))

path = np.array(path)
plt.plot(-path[:,1], -path[:,0])
plt.axis('equal')
plt.show()
```
